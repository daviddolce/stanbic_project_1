# How Many days has each BA worked

```python
df['Jina la BA:'] = df['Jina la BA:'].str.upper()
df['Jina la BA:'] = df['Jina la BA:'].str.strip()

df['Jina la BA:'] = df['Jina la BA:'].str.replace(r'\s[A-Z]+\s', ' ', regex=True)
```

```python
df['Jina la BA:'] = df['Jina la BA:'].str.replace(r'[J.]', '', regex=True)
#ba_df.ba_name.str.split(r'([A-Z][,])\s([A-Z]+)\s')
df['Jina la BA:'] =df['Jina la BA:'].str.replace(r'(GTEMU)', 'TEMU', regex=True)
df['Jina la BA:'] =df['Jina la BA:'].str.replace(r'(ESTER)', 'ESTHER', regex=True)
df['Jina la BA:'] =df['Jina la BA:'].str.replace(r'(MARIETTA)', 'MARIETHA', regex=True)
df['Jina la BA:'] =df['Jina la BA:'].str.replace(r'(EDAWRD)', 'EDWARD', regex=True)
df['Jina la BA:'] =df['Jina la BA:'].str.replace(r'(DITOPILE)', 'DITTOPILE', regex=True)
df['Jina la BA:'] =df['Jina la BA:'].str.replace(r'(SADA)', 'SADDA', regex=True)
df['Jina la BA:'] =df['Jina la BA:'].str.replace(r'(ELIA)', 'ELIYA', regex=True)
df['Jina la BA:'] =df['Jina la BA:'].str.replace(r'(SALUM)', 'SALIM', regex=True)
```
